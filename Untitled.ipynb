```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

url = "bank-full.csv"
df = pd.read_csv(url, header=None,sep=";")
# Assigning the provided headers
headers = ["age","job","marital","education","default","balance","housing","loan","contact","day","month","duration","campaign","pdays","previous","poutcome","y"]
df.columns = headers

df.dropna(subset=['balance'], axis=0, inplace=True)

mean_balance = df['balance'].mean()
df['balance'].replace(np.nan, mean_balance, inplace=True)

df.head(1000)
```

```python
data_types = df.dtypes

# Display the data types for each column
data_types
```

```python
# Min-Max normalization scales the data to be within the 0 to 1 range

min_balance = df['balance'].min()
max_balance = df['balance'].max()

# Apply min-max normalization
df['balance_normalized'] = (df['balance'] - min_balance) / (max_balance - min_balance)

# Display the first few rows to confirm normalization
df[['balance', 'balance_normalized']].head()
```

```python
balance_categories = pd.cut(df['balance'], bins=[-float('inf'), 0, float('inf')], labels=['Negative', 'Positive'])

balance_counts = balance_categories.value_counts()

zero_balance_count = df[df['balance'] == 0].shape[0]

balance_counts['Zero'] = zero_balance_count

balance_counts
```

```python
#    
categories = ['Negative', 'Zero', 'Positive']
counts = [balance_counts['Negative'], balance_counts['Zero'], balance_counts['Positive']]

#  
plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color=['red', 'grey', 'green'])
plt.title('Distribution of Balance Categories')
plt.xlabel('Balance Category')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle='--', alpha=0.7)

#  
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(y=df['balance'], x=df['education'])

plt.title('Boxplot of Balance')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['age'], y=df['balance'])

plt.title('Scatter Plot of Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance')

plt.show()
```

```python
grouped_data = df.groupby('education')['balance'].mean().sort_values()

grouped_df = grouped_data.reset_index()

plt.figure(figsize=(8, 3))
sns.heatmap(grouped_df.set_index('education').T, cmap='YlGnBu', annot=True, fmt=".0f")

plt.title('Average Balance by Education Level')
plt.show()
```

```python

pivot_table = df.pivot_table(index='education', columns='marital', values='balance', aggfunc='mean')

pivot_table
```

```python
correlation = df[['age', 'balance']].corr(method='pearson')

plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='balance', data=df, line_kws={"color":"red"})

plt.title('Scatter Plot with Regression Line: Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance')

plt.annotate(f'Pearson Correlation: {correlation.iloc[0,1]:.2f}', xy=(0.05, 0.95), 
             xycoords='axes fraction', ha='left', va='top', color='red')
plt.show()

correlation
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

numeric_correlation_matrix = numeric_df.corr(method='pearson')

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

plt.title('Heatmap of Pearson Correlation Coefficients (Numeric Columns)')
plt.show()

numeric_correlation_matrix
```

```python
balance_by_marital = df.groupby('marital')['balance'].apply(list)

balances_divorced = balance_by_marital['divorced']
balances_married = balance_by_marital['married']
balances_single = balance_by_marital['single']

anova_result = stats.f_oneway(balances_divorced, balances_married, balances_single)

anova_result
```
